```python
morph = pymorphy2.MorphAnalyzer()
def normilize_str(s1: str):
    if isinstance(s1, str):
        return [_[0].normal_form for _ in list(map(morph.parse, s1.split()))]
    else:
        return ""
    
def sets_jaccar(s1: str, s2: str):
    set1 = set(normilize_str(s1))
    set2 = set(normilize_str(s2))
    return len(set1 & set2)/ max(len(set1 | set2), 1)

# df.text.apply(lambda x: sets_jaccar(x, df.loc[0, 'text']))
sets_jaccar(df.loc[149, 'text'], df.loc[156, 'text'])
```

```python
df.loc[ :50000, 'norm'] = df[:50000].text.apply(lambda x: " ".join(normilize_str(x)))
```
