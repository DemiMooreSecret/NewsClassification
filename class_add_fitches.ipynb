```python
import pandas as pd
import re

df = pd.read_excel("Data/posts (1).xlsx")
df = df[["id", "text"]]
df_class = pd.read_excel("Data/class_not_mapped1000.xlsx")
df_class = df_class.drop("Unnamed: 0", axis=1)
df = df.join(df_class.set_index('id'), on='id')
df
```

```python
# remove not mapped
df = df[( df['–ù–∞–∑–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏'] != "–ù–µ –∑–Ω–∞—é") & (df['–ù–∞–∑–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏'].notna() )].reset_index(drop = True)
```

```python
# text cleaner
df['text'] = df.text.apply(lambda x: re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', x) if isinstance(x, str) else "")

df
```

```python
# word filter
def checkWord(s: str, list_words):
    if not isinstance(s, str):
        return False
    for word in list_words:
        if word in s:
            return True
    return False

list_econom = ['—Å–±–µ—Ä','—Ü–± ','–±–∞–Ω–∫','–¥–µ–Ω–µ–∂–Ω—ã–µ –ø–µ—Ä–µ–≤–æ–¥—ã']
list_tech = ['—Ç–µ—Ö–Ω–æ–ª–æ–≥','–∏–Ω–∂–µ–Ω–µ—Ä','—Ä–æ–±–æ—Ç']
list_politic = ['–Ω–∞—Ç–æ ','–±–∞–π–¥–µ–Ω','–±–æ–µ–≤—ã–µ','—Ç–µ—Ä—Ä–æ—Ä–∏—Å—Ç','—à—Ç—É—Ä–º','–ø–æ–ª–∏—Ç–∏–∫','–ü—Ä–∏–≥–æ–∂–∏–Ω']
list_show = ['—ç—Å—Ç—Ä–∞–¥','–∫–æ–Ω—Ü–µ—Ä—Ç','–∏–≤–∞–Ω—É—à–∫–∏','–∑–≤–µ–∑–¥–∞']
list_fashion = ['fashion', '–º–æ–¥–∞', '—Å—Ç–∏–ª—å']
list_btc = ['bitcoin','btc']
list_travel = ['–±–∞–≥–∞–∂','–∞–≤–∏–∞–±–∏–ª–µ—Ç']
list_educ = ['—à–∫–æ–ª', '—à–∫–æ–ª–∞', '–æ–±—Ä–æ–∑–æ–≤–∞—Ç–µ–ª—å–Ω–∞—è', '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ']
list_enter = ['—Ñ—É—Ç–±–æ–ª', '–ø—Ä–∞–∑–¥–Ω–∏–∫', '—Ñ–µ—Å—Ç–∏–≤–∞–ª—å']
df['econom'] = df.text.apply(lambda x: checkWord(x, list_econom))
df['tech'] = df.text.apply(lambda x: checkWord(x, list_tech))
df['politic'] = df.text.apply(lambda x: checkWord(x, list_politic))
df['show'] = df.text.apply(lambda x: checkWord(x, list_show))
df['fashion'] = df.text.apply(lambda x: checkWord(x, list_fashion))
df['btc'] = df.text.apply(lambda x: checkWord(x, list_btc))
df['travel'] = df.text.apply(lambda x: checkWord(x, list_travel))
df['educ'] = df.text.apply(lambda x: checkWord(x, list_educ))
df['enter'] = df.text.apply(lambda x: checkWord(x, list_enter))
df
```
